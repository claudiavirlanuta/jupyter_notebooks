# Visualize a Pandas DataFrame with Seaborn

Whether you are just getting starting with visualizations in Python, or you've already had enough with Matplotlib and are looking for a more elegant and easier-to-use viz library, look no further! Here is a step-by-step guide to getting started with Seaborn for visualizing a pandas dataframe.

## Setup

```python
# import all necessary modules
import pandas as pd
import random
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
# create a dataframe and populate it with numeric data
df = pd.DataFrame()

df['x'] = random.sample(range(1, 100), 25)
df['y'] = random.sample(range(1, 100), 25)
```

```python
df.head()
```

## Simple Scatterplot

```python
sns.lmplot('x', 'y', data=df, fit_reg=False)
```

## Scatterplot with Fitted Regression Line

```python
sns.lmplot('x', 'y', data=df, fit_reg=True)
```

## Density Plot

```python
sns.kdeplot(df.y)
```

```python
sns.kdeplot(df.y, df.x)
```

```python
sns.distplot(df.x)
```

## Histogram

```python
plt.hist(df.x)
sns.rugplot(df.x)
```

## Boxplot

```python
sns.boxplot([df.x, df.y], orient='v')
```
